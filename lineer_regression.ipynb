```python
import numpy as np
import matplotlib.pyplot as plt

# Study hours (input) and exam grade (output)
np.random.seed(1)
X = np.random.rand(100) * 10  # 0-10 hours
y = 3 * X + 7 + np.random.randn(100)  # y = 3x + 7 + noise
```

```python
def linear_regression(X, y):
    n = len(X)
    mean_x = np.mean(X)
    mean_y = np.mean(y)

    # Slope (m) calculate
    numerator = np.sum((X - mean_x) * (y - mean_y))
    denominator = np.sum((X - mean_x)**2)
    m = numerator / denominator

    # Intercept (b) calculate
    b = mean_y - m * mean_x

    return m, b

m, b = linear_regression(X, y)
print("Slope (m):", m)
print("Intercept (b):", b)
```

```python
def predict(x, m, b):
    return m * x + b

y_pred = predict(X, m, b)
```

```python
plt.scatter(X, y, label="Data")
plt.plot(X, y_pred, color='red', label=f"Lineer Regression y = {m: .2f} * x + {b: .2f}")
plt.xlabel("X (Study hours)")
plt.ylabel("Y (Exam Grade)")
plt.title("My Linear Regression Model")
plt.legend()
plt.show()
```

```python
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

mse = mean_squared_error(y, y_pred)
print("Mean Squared Error:", mse)
```
